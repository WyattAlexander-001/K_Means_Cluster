```python
# Libs
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# Generate our dataset
#Number of samples
#Number of clusters (centers)
#Number of features
#Cluster std
#Random state
dataset = make_blobs(n_samples = 200,
                    centers = 4,
                    n_features = 2,
                    cluster_std = 1.6,
                    random_state= 50)
#Creates two lists in a a two item list. We only want the index[0], so we make a new variable to store it
print(dataset)
```

```python
points = dataset[0]
print(points)
```

```python
# Import kmeans
from sklearn.cluster import KMeans
```

```python
# Create a KMeans objects
#Set number of clusters
kmeans = KMeans(n_clusters=4)
```

```python
#Fits object to dataset
kmeans.fit(points)
```

```python
#Data not clustered
plt.scatter(dataset[0][:,0],dataset[0][:,1])
```

```python
#Printing Clusters
clusters = kmeans.cluster_centers_
print(clusters)
```

```python
y_km = kmeans.fit_predict(points)
y_km
```

```python
#Plot data 



plt.scatter(points[y_km == 0,0], points[y_km == 0,1], s = 50, color = 'red')
plt.scatter(points[y_km == 1,0], points[y_km == 1,1], s = 50, color = 'blue')
plt.scatter(points[y_km == 2,0], points[y_km == 2,1], s = 50, color = 'orange')
plt.scatter(points[y_km == 3,0], points[y_km == 3,1], s = 50, color = 'green')

plt.scatter(clusters[0][0], clusters[0][1], marker = '*', s=300, color = 'black')
plt.scatter(clusters[1][0], clusters[1][1], marker = '*', s=300, color = 'black')
plt.scatter(clusters[2][0], clusters[2][1], marker = '*', s=300, color = 'black')
plt.scatter(clusters[3][0], clusters[3][1], marker = '*', s=300, color = 'black')
```
